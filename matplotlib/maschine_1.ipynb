```python
import matplotlib.pyplot as plt

bm_large = [0, 0.099, 0.174, 0.248, 0.322, 0.399, 0.466, 0.538, 0.598, 0.646, 0.697]
h_large = [0, 78.512, 157.025, 235.537, 314.050, 392.562, 471.074, 549.587, 628.099, 706.612, 785.124]

# Choose two points to calculate dx and dy
i1, i2 = 2, 3  # Indices of the two points to compare
x1, x2 = h_large[i1], h_large[i2]
y1, y2 = bm_large[i1], bm_large[i2]
dx = x2 - x1
dy = y2 - y1

fig, ax = plt.subplots()
ax.set_xlabel("Magnetic field strength (A/m)")
ax.set_ylabel("Magnetic flux density (T)")
ax.set_xlim(xmin=h_large[0], xmax=round(h_large[-1], -2))
ax.set_ylim(ymin=bm_large[0], ymax=round(bm_large[-1], 1))
ax.set_title("B vs H for larger core")
ax.grid(True)

ax.arrow(x1, y1, dx, 0, length_includes_head=True)
ax.arrow(x2, y1, 0, dy, length_includes_head=True)

ax.text(x1 + dx / 2, y1 - 0.05, f"$dx = {dx:.1f}$", fontsize=12, ha="center")
ax.text(x2 + 20, y1 + dy / 2, f"$dy = {dy:.3f}$", fontsize=12, va="center")

ax.plot(h_large, bm_large, color="tab:blue")
ax.plot([x1, x2], [y1, y2], 'o')

plt.show()
```
