```python
# First bit of matplotlib
# Plotting discrete values
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4]
y_values = [5, 4, 6, 2]

plt.plot(x_values, y_values)
```

```python
# Plotting a function
import numpy as np

x = np.linspace(1, 100, 500)
y = np.log(x)

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
# Plotting a function
x = np.linspace(1, 2 * np.pi, 100)
y = np.sin(x) * np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y)

fig.set_size_inches(6, 4)
fig.set_dpi(300)
plt.savefig("sin_times_cosine.png")
```

```python
# Labels, titles, et cetera
from matplotlib.ticker import MultipleLocator

x = np.linspace(1, 20, 50)
y = x**2

fig, ax = plt.subplots()
ax.set_xlabel("Inputs")
ax.set_ylabel("Time (s)")
ax.set_title("Time Complexity of Bubblesort")

# These two lines are used to keep the x-axis major tick labels limited to natural numbers
ax.xaxis.set_major_locator(MultipleLocator(1))

ax.plot(x, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(444)

rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots()
ax.stackplot(yrs, rng + rnd, labels=["East Asia", "Eurasia", "Oceania"])
ax.set_title("Combined debt growth over time")
ax.legend(loc="upper left")
ax.set_ylabel("Total debt")
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
```
